```python
# 4.How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import seaborn as sns
import matplotlib.pyplot as plt
iris = sns.load_dataset('iris')
numeric_iris = iris.select_dtypes(include=['float64', 'int64'])
correlation_matrix = numeric_iris.corr()
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt='.2f')
plt.title('Correlation Matrix Heatmap')
plt.show()
```

```python
# 5. Generate a bar plot using Plotly
import plotly.graph_objects as go
categories = ['Category A', 'Category B', 'Category C', 'Category D']
values = [10, 15, 7, 12]
fig = go.Figure(data=[go.Bar(x=categories, y=values, marker_color='royalblue')])
fig.update_layout(
    title='Bar Plot Example',
    xaxis_title='Categories',
    yaxis_title='Values',
    template='plotly_white'
)
fig.show()
```

```python
# 8. Create a line plot with multiple lines using Matplotlib
import pandas as pd
import numpy as np
x=np.linspace(0,10,50)
y1=np.sin(x)
y2=np.cos(x)
import matplotlib.pyplot as plt
plt.plot(x, y1, label='sin(x)')
plt.plot(x, y2, label='cos(x)')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Multiple Lines Plot')
plt.legend()
plt.show()

```

```python
# 10.Create a histogram using Seaborn to visualize a distribution
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
data = np.random.randn(1000)
sns.histplot(data, kde=True, color='blue', bins=30)
plt.title('Histogram with Seaborn')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

```python
# 11. Perform matrix multiplication using NumPy
arr1=np.array([1,2,3,4])
arr2=np.array([5,6,7,8])
result=np.matrix(arr1).T*np.matrix(arr2)
print(result)
```
